```python
import pandas as pd

# Load Excel file
file_path = "GDP.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
print(df.columns)
```

```python
# 1. Melt the DataFrame to transform it from wide to long format
df_long = pd.melt(df, 
                  id_vars=['Country Name', 'Country Code'], 
                  var_name='Time', 
                  value_name='GDP')
```

```python
# 2. Rename columns to match your desired output
df_long = df_long.rename(columns={'Country Code': 'ISO2_code'})
```

```python
# 3. Sort the DataFrame by country and then by year
df_long = df_long.sort_values(by=['ISO2_code', 'Time'])
```

```python
# 4. (Optional) Remove the 'Country Name' if you don't need it
df_long = df_long.drop(columns=['Country Name'])


```

```python
# 5. (Optional) Reset the index
df_long = df_long.reset_index(drop=True)

# Display the first 10 rows of the transformed data
print(df_long.head(50))
```

```python
# Save the DataFrame to a CSV file
df_long.to_csv('gdp_data.csv', index=False)
```
